```python
def get_data(statdataID,startPosition):
      # APIエンドポイントのURL
    url = 'http://api.e-stat.go.jp/rest/3.0/app/getSimpleStatsData'
    # パラメータの設定
    params = {
        'appId': api_key,
        'lang': 'E',
        'statsDataId': statdataID,
        'metaGetFlg': 'N',
        'cntGetFlg': 'N',
        'explanationGetFlg': 'N',
        'annotationGetFlg': 'N',
        'sectionHeaderFlg': '2',
        'replaceSpChars': '0',
        'startPosition':str(startPosition),
        #'cdCat01':'480300100',
        'cdCat02':'160,170,190,200,220,230,250,260,280,290,310,320,340,350,370,380,400,410,430,440,460,470,490,500',
        #'cdCat03':'480300100',
        #'cdArea':'50224'
    }
    # APIにリクエストを送信
    response = requests.get(url, params=params)
    # レスポンスのCSVデータを読み込む
    csv_data = response.text
    # CSVデータをパースしてDataFrameに変換
    df = pd.read_csv(StringIO('\n'.join(csv_data.split('\n')[0:])))
    return df
```

```python
# セルの実行開始時刻を記録
start_time = time.time()
# ここに実行したいコードを記述
startPosition = 0
all_dfs = []

for _ in range(99999):  # 任意の大きな範囲を指定
    df = get_data(statdataID, startPosition)
    print(len(df))

    if len(df) >= 99999:
        print('まだある', startPosition)
        all_dfs.append(df)
        startPosition += len(df) + 1

    if len(df) < 99999:
        all_dfs.append(df)
        print('終了：', startPosition)
        break  # 取得したデータが 100000 行未満なら終了
# すべての DataFrame を縦結合
final_df = pd.concat(all_dfs, ignore_index=True)
# セルの実行終了時刻を記録
end_time = time.time()
# 経過時間を計算してプリント
execution_time = end_time - start_time
print(f"セルの実行時間: {execution_time}秒")
```
